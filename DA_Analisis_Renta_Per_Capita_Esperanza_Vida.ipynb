# Análisis renta per cápita en relación a la esperanza de vida


## 1. Importar librerías



```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
```

## 2. Lectura de los datos

```python
# Cargo los datos
uploaded = files.upload()
```

```python
# Leo los datos
df = pd.read_csv('Info_pais.csv',
    encoding='ISO-8859-1',
    delimiter= ';',
    decimal=','
)

df.head()
```

```python
# Ordeno por población
df_order = df.sort_values(
    'Poblacion',
    ascending=False
)

df_order.head()
```

## 3. Tratamiento y visualización de los datos

```python
# Hago un primer gráfico de referencia
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida');
```

```python
# Normalizo la población
df_order['Poblacion_normalizada'] = df_order['Poblacion']/(max(df_order['Poblacion'])/10000)
df_order.head()
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada'],
    c = df_order['Poblacion_normalizada'],
    alpha = 0.5
)

plt.title('Renta per cápita vs Esperanza de vida')
plt.xlabel('Renta per cápita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5,10)

df_order.reset_index(drop=True, inplace=True)

for i in range (0,10):
    plt.annotate(
    df_order['País'][i],
    (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )
```

## 4. Conclusión final

Según el gráfico de arriba que compara la renta per cápita de un país respecto a su esperanza de vida, podemos llegar a la conclusión de que aquellos países con más renta pér cápita son los que más esperanza de vida tienen y viceversa, aquellos con una baja renta per cápita cuentan con una baja esperanza de vida. Esto nos indica que existe una correlación clara entre ambas variables y si quisiéramos mejorar la esperanza de vida, deberíamos mejorar antes la renta per cápita.
